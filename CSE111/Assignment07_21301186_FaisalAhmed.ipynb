```python
#Task-2
class Vehicle:
    def __init__(self):
        self.x = 0
        self.y = 0
    def moveUp(self):
        self.y+=1
    def moveDown(self):
        self.y-=1
    def moveRight(self):
        self.x+=1
    def moveLeft(self):
        self.x-=1
    def __str__(self):
        return '('+str(self.x)+' , '+str(self.y)+')'
class Vehicle2010(Vehicle):
    def moveUpperRight(self):
        Vehicle.moveUp(self)
        Vehicle.moveRight(self)
    def moveUpperLeft(self):
        Vehicle.moveUp(self)
        Vehicle.moveLeft(self)
    def moveLowerRight(self):
        Vehicle.moveDown(self)
        Vehicle.moveRight(self)
    def moveLowerLeft(self):
        Vehicle.moveDown(self)
        Vehicle.moveLeft(self)
    def equals(self,other):
        if self.x == other.x and self.y == other.y:
            return True
        else:
            return False
        
        
print('Part 1')
print('------')
car = Vehicle()
print(car)
car.moveUp()
print(car)
car.moveLeft()
print(car)
car.moveDown()
print(car)
car.moveRight()
print(car)
print('------')
print('Part 2')
print('------')
car1 = Vehicle2010()
print(car1)
car1.moveLowerLeft()
print(car1)
car2 = Vehicle2010()
car2.moveLeft()
print(car1.equals(car2))
car2.moveDown()
print(car1.equals(car2))
```

```python
#Task-3
class Tournament:
    def __init__(self,name="Default"):
        self.__name = name
    def set_name(self,name):
        self.__name = name
    def get_name(self):
        return self.__name
class Cricket_Tournament(Tournament):
    def __init__(self,new_name="Default",team_number=0,Type="No Type"):
        Tournament.__init__(self,new_name)
        self.team_number = team_number
        self.type = Type
    def detail(self):
        return f"Cricket Tournament Name: {self.get_name()}\nNumber of Teams: {self.team_number}\nType: {self.type}"
class Tennis_Tournament(Tournament):
    def __init__(self,name,player_numbers):
        Tournament.__init__(self,name)
        self.player_numbers = player_numbers
    def detail(self):
        return  f"Tennis Tournament Name: {self.get_name()}\nNumber of Players: {self.player_numbers}"
        
ct1 = Cricket_Tournament()
print(ct1.detail())
print("-----------------------")
ct2 = Cricket_Tournament("IPL",10,"t20")
print(ct2.detail())
print("-----------------------")
tt = Tennis_Tournament("Roland Garros",128)
print(tt.detail())
```

```python
#Task-4
class Product:
    def __init__(self,id, title, price):
        self.__id = id
        self. __title = title
        self. __price = price
    def get_id_title_price(self):
        return "ID: "+str(self.__id)+" Title: "+self.__title+ " Price: "+str(self.__price)
class Book(Product):
    def __init__(self,id,title,price,isbn,publisher):
        Product.__init__(self,id, title, price)
        self.isbn = isbn
        self.publisher = publisher
    def printDetail(self):
        return f"{self.get_id_title_price()}\nISBN: {self.isbn} Publisher: {self.publisher}"
class CD(Product):
    def __init__(self,id,title,price,band,duration,genre):
        Product.__init__(self,id, title, price)
        self.band = band
        self.duration = duration
        self.genre = genre
    def printDetail(self):
        return f"{self.get_id_title_price()}\nBand: {self.band} Duration: {self.duration}minutes\nGenre: {self.genre}"
book = Book(1,"The Alchemist",500,"97806","HarperCollins")
print(book.printDetail())
print("-----------------------")
cd = CD(2,"Shotto",300,"Warfaze",50,"Hard Rock")
print(cd.printDetail())
```

```python
#Task-5
class Animal: 
    def __init__(self,sound):
        self.__sound = sound
 
    def makeSound(self):
        return self.__sound
class Printer: 
    def printSound(self, a):
        print(a.makeSound())
class Dog(Animal):
    def __init__(self,sound):
        Animal.__init__(self,sound)
class Cat(Animal):
    def __init(self,sound):
        Animal.__init__(self,sound)

d1 = Dog('bark')
c1 = Cat('meow')
a1 = Animal('Animal does not make sound')
pr = Printer()
pr.printSound(a1)
pr.printSound(c1)
pr.printSound(d1)
```

```python
#Task-6
class Shape:
    def __init__(self, name='Default', height=0, base=0):
        self.area = 0
        self.name = name
        self.height = height
        self.base = base
    def get_height_base(self):
        return "Height: "+str(self.height)+", Base: "+str(self.base)
class triangle(Shape):
    def __init__(self, name="Triangle", height=0, base=0):
        Shape.__init__(self, name, height, base)
    def calcArea(self):
        self.area += (0.5*self.base*self.height)
    def printDetail(self):
        return f"Shape name: {self.name}\n{self.get_height_base()}\nArea: {self.area}"
class trapezoid(Shape):
    def __init__(self, name="Triangle", height=0, base=0, one_side=0):
        Shape.__init__(self, name, height, base)
        self.one_side = one_side
    def calcArea(self):
        self.area += (((self.one_side + self.base)/2) * self.height)
    def printDetail(self):
        return f"Shape name: {self.name}  \n{self.get_height_base()}, Side_A: {self.one_side} \nArea: {self.area}"
        
        
        
        

tri_default = triangle()
tri_default.calcArea()
print(tri_default.printDetail())
print('--------------------------')
tri = triangle('Triangle', 10, 5)
tri.calcArea()
print(tri.printDetail())
print('---------------------------')
trap = trapezoid('Trapezoid', 10, 6, 4)
trap.calcArea()
print(trap.printDetail())
```

```python
#Task-7
class Football:
    def __init__(self, team_name, name, role):
        self.__team = team_name
        self.__name = name
        self.role = role
        self.earning_per_match = 0
    def get_name_team(self):
        return 'Name: '+self.__name+', Team Name: ' +self.__team
class Player(Football):
    def __init__(self,team_name, name, role, goal_number=0, match_played=0):
        Football.__init__(self, team_name, name, role)
        self.goal_number = goal_number
        self.match_played = match_played
        self.ratio = 0
    def calculate_ratio(self):
        self.earning_per_match += ((self.goal_number * 1000) + (self.match_played * 10))
        self.ratio += (self.goal_number / self.match_played)
    def print_details(self):
        print(f"{self.get_name_team()}\nTeam Role: {self.role}\nTotal Goal: {self.goal_number}, Total Played: {self.match_played}\nGoal Ratio: {self.ratio}\nMatch Earning: {self.earning_per_match}K")
class Manager(Football):
    def __init__(self,team_name, name, role, matches_win):
        Football.__init__(self, team_name, name, role)
        self.matches_win = matches_win
        self.earning_per_match += (self.matches_win * 1000)
    def print_details(self):
        print(f"{self.get_name_team()}\nTeam Role: {self.role}\nTotal Win: {self.matches_win}\nMatch Earning: {self.earning_per_match}K")

player_one = Player('Juventus', 'Ronaldo', 'Striker', 25, 32)
player_one.calculate_ratio()
player_one.print_details()
print('------------------------------------------')
manager_one = Manager('Real Madrid', 'Zidane', 'Manager', 25)
manager_one.print_details()
```
