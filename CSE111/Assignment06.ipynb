```python
#Task-5
class Employee:
    def __init__(self,name,workingPeriod):
        self.name = name
        self.workingPeriod = workingPeriod
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        return Employee(name,2021-year)
    @staticmethod
    def experienceCheck(workingperiod,gender):
        if workingperiod >= 3:
            if gender == "male":
                return "He is experienced"
            else:
                return "She is experienced"
        else:
            if gender == "male":
                return "He is not experienced"
            else:
                return "She is not experienced"
            
            
employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#Task-6
class Laptop:
    laptopCount = 0
    def __init__(self,name,count):
        self.name = name
        self.count = count
        Laptop.laptopCount += int(count)
    @staticmethod
    def advantage():
        print("Laptops are portable")
    @classmethod
    def resetCount(cls):
        Laptop.laptopCount = 0
lenovo = Laptop("Lenovo", 5);
dell = Laptop("Dell", 7);
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount)
```

```python
#Task-7
class Cat:
    Number_of_cats = 0
    def __init__(self,color,position):
        self.color = color
        self.position = position
        Cat.Number_of_cats += 1
    @classmethod
    def no_parameter(cls):
        return cls("white","sitting")
    @classmethod
    def first_parameter(cls,color):
        return cls(color,"sitting")
    @classmethod
    def second_parameter(cls,position):
        return cls("Grey",position)
    
    def changeColor(self,new_color):
        self.color = new_color
    def printCat(self):
        print(f"{self.color} cat is {self.position}")
        
        
print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#Task-8
class Cylinder:
    radius = 5
    height = 18
    def __init__(self,radius,height):
        print(f"Detault radius = {Cylinder.radius} and height = {Cylinder.height}")
        Cylinder.radius = radius
        Cylinder.height = height
        print(f"Updated radius = {Cylinder.radius} and height = {Cylinder.height}")
    @classmethod
    def changeFormat(cls, string):
        listt = string.split("-")
        for i in range(len(listt)):
            if i == 0:
                radius = listt[i]
            else:
                height = listt[i]
        solve = Cylinder(radius, height)
        return solve
    @classmethod
    def swap(cls,radius,height):
        solve = Cylinder(height,radius)
        return solve
    @staticmethod
    def area(radius, height):
        radius = float(radius)
        height = float(height)
        area = (2 * 3.141592653589793 * radius ** 2) + (2 * 3.141592653589793 * radius * height)
        print("Area:", area)
    @staticmethod
    def volume(radius, height):
        radius = float(radius)
        height= float(height) 
        volume = (3.141592653589793 * height * radius ** 2)
        print("Volume:", volume)

    
c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#Task-10
class SultansDine:
    branchnumber = 0
    totalSell=0
    food_list = []
    def __init__(self,branch):
        self.branch = branch
        SultansDine.branchnumber += 1
        self.Branch_sell = 0

    def sellQuantity(self,amount):
        if amount < 10:
            self.Branch_sell = amount * 300
        elif amount < 20:
            self.Branch_sell = amount * 350
        else:
            self.Branch_sell = amount * 400
        SultansDine.totalSell += self.Branch_sell
        SultansDine.food_list.append(self)
    def branchInformation(self):
        print(f"Branch Name: {self.branch} \nBranch sell: {self.Branch_sell}")

    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.branchnumber} \nTotal sell: {cls.totalSell} Taka")
        for i in cls.food_list:
            print(f"Branch Name: {i.branch}, Branch Sell: {i.Branch_sell} Taka \nBranch consists of total sell\'s: {((i.Branch_sell / cls.totalSell) * 100):.2f} %")


SultansDine.details()
print('########################')
dhanmodi = SultansDine('Dhanmondi')
dhanmodi.sellQuantity(25)
dhanmodi.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()
print('========================')
gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
